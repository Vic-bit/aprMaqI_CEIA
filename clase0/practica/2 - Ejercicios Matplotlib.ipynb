Ejercicio 5:

Cree un gráfico en estructura de dos por dos usando **figure** para las siguientes funciones hiperbólicas en el dominio $ -1 \leq x \leq 1$

| Estructura del grafico | |
| ---------------------  | -- |
| $senh(x) \; vs \; x$ | $cosh(x) \; vs \; x$ |
| $tanh(x) \; vs \; x$ | $sech(x) \; vs \; x$ |

```python
x = np.linspace(-1,1)
fig = plt.figure()

senh = plt.subplot(221)
senh.plot(x,np.sinh(x))
senh.set_xlabel('x')
senh.set_ylabel('senh(x)')
senh.grid()

cosh = plt.subplot(222)
cosh.plot(x,np.cosh(x))
cosh.set_xlabel('x')
cosh.set_ylabel('cosh(x)')
cosh.grid()

tanh = plt.subplot(223)
tanh.plot(x,np.tanh(x))
tanh.set_xlabel('x')
tanh.set_ylabel('tanh(x)')
tanh.grid()

sech = plt.subplot(224)
sech.plot(x,1/np.cosh(x))
sech.set_xlabel('x')
sech.set_ylabel('sech(x)')
sech.grid()

fig.subplots_adjust(wspace=0.5)
```
